```python
def run_scored(batsman_name):
    vk = delivery[delivery['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

<h1>The merge() Function</h1>

<p>
The merge() function in pandas is used to combine two DataFrames based on common columns or indices, similar to SQL joins.
</p>

<ul>
  <li><strong>inner</strong> â€“ Only matching rows (default)</li>
  <li><strong>outer</strong> â€“ All rows (union), fill with NaN</li>
  <li><strong>left</strong> â€“ All rows from left, matched from right</li>
  <li><strong>right</strong> â€“ All rows from right, matched from left</li>
</ul>
